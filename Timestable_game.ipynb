```python

#CURRENT!!
#create a timestable quiz using python and tkinter GUI

#import libraries 
import tkinter as tk
from tkinter import *

import random 

#assign variables 
a=0
b=0
q_number=0
score=0
turns=0

#create a function to welcome the user
def welcome_msg():
    question.set('Hi! Lets practise our 1 to 10 timestables! Click the start button to begin.')
    ent_answer.grid_remove()
    submit_ans_btn.grid_remove()
    next_btn.grid_remove()
    yes_btn.grid_remove()
    no_btn.grid_remove()

#create a function to start the quiz
def start_quiz(): 
        start_quiz_btn.grid_remove()
        ent_answer.grid()
        next_btn.grid()
        submit_ans_btn.grid()
        next_btn.config(state='disabled')
        next_btn_text.set('Next')
        answer.config(text='')
        global a
        global b
        global q_number 
        global turns 
        turns=0
        q_number=0
        q_number=q_number+1
        a=random.randint(0,10) 
        b=random.randint(0,10)
        question.set('What is ' + str(a) + ' x ' + str(b) + ' = ?')
        print('This is question, ' + str(q_number))

#create a function to submit and check the user's answer 
def submit_answer():

    try: 
        int(ent_answer.get())

        correct_answer = a*b 
        global score 
        global turns
            
        if int(ent_answer.get()) == correct_answer: 
            turns=turns+1
            score=score+1
            answer.config(text='Correct!')
            ent_answer.delete(0,END)
            ent_answer.config(state='disabled')
            submit_ans_btn.config(state='disabled')
            next_btn.config(state='normal')
            print('you scored, ' +str(score) +' out of ' + str(q_number))
           
            
        else:
            turns=turns+1
            if turns == 1:
                answer.config(text='Incorrect! You have one more try')
                ent_answer.delete(0,END)
                next_btn.config(state='disabled')
                print(turns)
            elif turns >1 :
                answer.config(text='The correct answer is ' + str(a*b) +'.')
                ent_answer.delete(0,END)
                ent_answer.config(state='disabled')
                submit_ans_btn.config(state='disabled')
                next_btn.config(state='normal')
                
                print(turns)

    except: 
        answer.config(text='Please enter an integer.')
        ent_answer.delete(0,END)
        print(turns)
    
#create a function to move on to the next question 
def next_question():
        next_btn.config(state='disabled')
        answer.config(text='')
        global a
        global b
        global q_number 
        global turns 
        global score 
        turns=0
        
        q_number=q_number+1
        #score=score+1
        a=random.randint(0,10) 
        b=random.randint(0,10)
        question.set('What is ' + str(a) + ' x ' + str(b) + ' = ?')
        #delete data input
        ent_answer.delete(0,END)
        next_btn_text.set('Next')
        
        if q_number == 5:
            next_btn_text.set('Finish')
            next_btn.config(command=finish_quiz)
            print('This is question, ' + str(q_number))      

        if q_number > 5:
            question.set('Well done, you have completed the quiz! You scored, ' +str(score) + ' out of ' + str(q_number) + '. Would you like to play again?')
            next_btn_text.set('Finish')
            yes_btn.grid()
            no_btn.grid()
            print('This is question, ' + str(q_number))
           
            submit_ans_btn.grid_remove()
            next_btn.grid_remove()
            ent_answer.grid_remove()
            
        ent_answer.config(state='normal')
        submit_ans_btn.config(state='normal')    
        #next_btn.config(state='normal')    

#create a function to finish the quiz with the option to play again        
def finish_quiz():
    global score 
    global q_number
    
    question.set('Well done, you have completed the quiz! You scored, ' +str(score) + ' out of ' + str(q_number) + '. Would you like to play again?')
    yes_btn.grid()
    no_btn.grid()
    submit_ans_btn.grid_remove()
    next_btn.grid_remove()
    ent_answer.grid_remove()
    
    answer.config(text='')

#create a function to play the quiz again     
def play_again():
        
        global a
        global b
        global q_number 
        global turns 
        global score 
        q_number=0
        turns=0    
        score=0  
        start_quiz_btn.grid_remove()
        yes_btn.grid_remove()
        no_btn.grid_remove()
        ent_answer.grid()
        ent_answer.config(state='normal')
        next_btn.grid()
        next_btn.config(state='disabled')
        submit_ans_btn.grid()
        submit_ans_btn.config(state='normal')
        #next_btn.config(state='disabled')
        next_btn_text.set('Next')
        next_btn.config(command=next_question)
        answer.config(text='')
        q_number=q_number+1
        a=random.randint(0,10) 
        b=random.randint(0,10)
        question.set('What is ' + str(a) + ' x ' + str(b) + ' = ?')
        print('This is question, ' + str(q_number))
        
#create a function to thank the user and end the quiz         
def end_quiz():
    question.set('Thank you for playing. You may close the window now.')
    ent_answer.grid_remove()
    submit_ans_btn.grid_remove()
    next_btn.grid_remove()
    yes_btn.grid_remove()
    no_btn.grid_remove()
    answer.config(text='')
    
    ent_answer.config(state='normal')
    submit_ans_btn.config(state='normal')  
    
#create a tkinter GUI 
window=tk.Tk()
window.title('Lets practise our times tables!')

#assign string var to variables to enable changes to text
question=StringVar()
next_btn_text=StringVar()


#display a question using a label widget 
Question_label=tk.Label(window, textvariable=question)
Question_label.grid(row=1, column=0)


#create a text box for user input 
ent_answer=tk.Entry(width=30)
ent_answer.grid(row=1, column=1, padx=20)

#create a start quiz button
start_quiz_btn=tk.Button(text='Start', command=start_quiz)
start_quiz_btn.grid(row=3, column=3)

#create a submit answer button 
submit_ans_btn=tk.Button(text='Submit Answer', command=submit_answer)
submit_ans_btn.grid(row=3, column=1)

#create a next question button
next_btn=tk.Button(textvariable=next_btn_text, command=next_question)
next_btn.grid(row=3, column=2)

#create a yes play again button 
yes_btn=tk.Button(text='Yes', command=play_again)
yes_btn.grid(row=4, column=1)

#create a no button 
no_btn=tk.Button(text='No', command=end_quiz)
no_btn.grid(row=4, column=2)

#create an answer in the GUI to display output messages 
answer=tk.Label(text='')
answer.grid(row=5, column=1)

#call the first function 
welcome_msg()

#exit GUI
window.mainloop()
window.quit()
```
